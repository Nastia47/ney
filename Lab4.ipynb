```python
#вирішити завдання класифікації для набору даних Fashion MNIST
#Варіант5: 4, 7, 1
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.datasets import fashion_mnist

(train_data, train_labels), (test_data, test_labels) = fashion_mnist.load_data()
train_data_choose = []
train_labels_choose = []
for n in range(len(train_data)):
  if(train_labels[n] == 4 or train_labels[n] == 7 or train_labels[n] == 1):
    train_data_choose.append(train_data[n])
    train_labels_choose.append(train_labels[n])
test_data_choose = []
test_labels_choose = []
for n in range(len(test_data)):
  if(test_labels[n] == 4 or test_labels[n] == 7 or test_labels[n] == 1):
    test_data_choose.append(test_data[n])
    test_labels_choose.append(test_labels[n])
train_labels1=np.array(train_labels_choose)   
train_data1=np.array(train_data_choose)
test_data1=np.array(test_data_choose)
test_labels1=np.array(test_labels_choose)
#normalization
train_data = train_data / 255.0
test_data = test_data / 255.0
train_data.min(), train_data.max()
tf.random.set_seed(42)
#model
model_1 = tf.keras.Sequential([
  tf.keras.layers.Flatten(input_shape=(28, 28)),
  tf.keras.layers.Dense(4, activation="relu"),
  tf.keras.layers.Dense(4, activation="relu"),
  tf.keras.layers.Dense(10, activation="softmax") 
])
model_1.compile(loss=tf.keras.losses.SparseCategoricalCrossentropy(),
                 optimizer=tf.keras.optimizers.Adam(),
                 metrics=["accuracy"])
lr_scheduler = tf.keras.callbacks.LearningRateScheduler(lambda epoch: 1e-3 * 10**(epoch/20))

# Fit the model
find_lr_history = model_1.fit(train_data1,
                               train_labels1,
                               epochs=40,
                               validation_data=(test_data1, test_labels1),
                               callbacks=[lr_scheduler])
```
